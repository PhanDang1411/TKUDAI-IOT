# Lab 4 Standalone Runner

Use this notebook as the main Kaggle notebook.
Required inputs:
1. `lab4_project_for_kaggle_safe.zip`
2. `energydata_complete.csv` (UCI) (optional, fallback sample is included)

```python
import os, glob, zipfile, pathlib

print('Kaggle input roots:', os.listdir('/kaggle/input'))
zip_candidates = glob.glob('/kaggle/input/**/lab4_project_for_kaggle_safe.zip', recursive=True)
if not zip_candidates:
    zip_candidates = glob.glob('/kaggle/input/**/*.zip', recursive=True)

if not zip_candidates:
    raise FileNotFoundError('Cannot find project zip in /kaggle/input')

project_zip = zip_candidates[0]
print('Project zip:', project_zip)

dest = '/kaggle/working/lab4'
pathlib.Path(dest).mkdir(parents=True, exist_ok=True)
with zipfile.ZipFile(project_zip, 'r') as z:
    z.extractall(dest)

print('Extracted to:', dest)
print('Top-level:', os.listdir(dest)[:20])
```

```python
import os, glob, shutil

requirements = glob.glob('/kaggle/working/lab4/**/requirements.txt', recursive=True)
if not requirements:
    raise FileNotFoundError('requirements.txt not found after extraction')

project_dir = os.path.dirname(requirements[0])
os.chdir(project_dir)
print('Project dir:', project_dir)

dataset_candidates = glob.glob('/kaggle/input/**/energydata_complete.csv', recursive=True)
if dataset_candidates:
    target = os.path.join(project_dir, 'data', 'energydata_complete.csv')
    shutil.copyfile(dataset_candidates[0], target)
    print('Copied UCI dataset to:', target)
else:
    print('No UCI dataset uploaded. Will use fallback sample in data/.')

print('Project files:', os.listdir(project_dir)[:30])
```
